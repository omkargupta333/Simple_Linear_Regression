```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canadapci.csv')
df.head(5)
```

```python
plt.scatter(df['year'], df['pci'])
plt.xlabel('Year', fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.plot
```

```python
plt.scatter(df['year'], df['pci'], color = 'red', marker='+')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df.pci)
```

```python
m = reg.coef_    # slop(m)
```

```python
b = reg.intercept_   #intercept(b)
```

```python
# reg.predict(2023)
```

```python
plt.xlabel('year', fontsize=20)
plt.ylabel('pci', fontsize=20)
plt.scatter(df.year, df.pci, color='red', marker='+')
plt.plot(df.year,reg.predict(df[['year']]), color='blue')
```

```python
def predict(x):
    x = m*x+b
    return x
```

```python
predict(2023)
```
